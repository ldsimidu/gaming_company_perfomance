```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tests_perfomance = "spreadsheet/tests_per_team.xlsx"
df = pd.read_excel(tests_perfomance, sheet_name='Sheet1')

plt.figure(figsize=(12, 7))

for i, column in enumerate(df.columns[1:]):
    marker = ['o', 's', 'D', '^', 'v', '*'][i]
    plt.plot(df['Sprint'], df[column], marker=marker, label=column, linewidth=3)

plt.title('Tendência de Testes Falhados por Sprint', fontsize=14)
plt.xlabel('Sprints', fontsize=12)
plt.ylabel('Testes Falhados', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
cores = [
    '#1f77b4', '#ff7f0e', '#2ca02c', 
    '#d62728', '#9467bd', '#8c564b'
]

s1_data = df[df['Sprint'] == 'S1'].iloc[0]
times = s1_data.index[1:] 
valores = s1_data.values[1:]

plt.figure(figsize=(14, 7))

bars = plt.bar(times, valores, color=cores, alpha=0.7)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=12)

plt.title('Testes Falhados na Sprint S1 por Time', fontsize=16, pad=20)
plt.xlabel('Times', fontsize=12)
plt.ylabel('Quantidade de Testes Falhados', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()

plt.show()
```

```python
s1_data = df[df['Sprint'] == 'S2'].iloc[0]
times = s1_data.index[1:] 
valores = s1_data.values[1:]

cores = [
    '#1f77b4', '#ff7f0e', '#2ca02c', 
    '#d62728', '#9467bd', '#8c564b'
]

plt.figure(figsize=(14, 7))

bars = plt.bar(times, valores, color=cores, alpha=0.7)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', fontsize=12)

plt.title('Testes Falhados na Sprint S2 por Time', fontsize=16, pad=20)
plt.xlabel('Times', fontsize=12)
plt.ylabel('Quantidade de Testes Falhados', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()

plt.show()
```

```python
sprints = df.iloc[:, 0].tolist()
teams = df.columns[1:]
data = [df[team].tolist() for team in teams]

x = np.arange(len(sprints))
width = 0.1

fig, ax = plt.subplots(figsize=(12, 7))

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']

rects = []
for i, (team, values, color) in enumerate(zip(teams, data, colors)):
    offset = width * (i - len(teams)/2 + 0.5)
    rects.append(ax.bar(x + offset, values, width, label=team, color=color))

ax.set_xlabel('Sprints', fontsize=12)
ax.set_ylabel('Testes Falhados', fontsize=12)
ax.set_title('Progressão de Testes Falhados por Time e Sprint', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(sprints)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True, linestyle='--', alpha=0.7)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

for r in rects:
    autolabel(r)

plt.tight_layout()
plt.show()
```
